```python
#print len(pred)
#print len(y_test)
#print y_test.shape
#print y_test
differences = list()
for q in range(len(y_test)):
    differences.append(y_test[q]-pred[q])
acc0 = 0
for w in differences:
    if w == 0:
        acc0 += 1
print "Completely Accurate",(acc0/float(len(pred)))*100, "%"
acc1 = 0
for w in differences:
    if w == -1:
        acc1 += 1
print "One Above",(acc1/float(len(pred)))*100, "%"
acc2 = 0
for w in differences:
    if w == 1:
        acc2 += 1
print "One Below",(acc2/float(len(pred)))*100, "%"
print "Sum of the Three", ((acc0+acc1+acc2)/float(len(pred)))*100, "%"
```

We can predict within +/- 1 quartile at least 60% of the time, which means that generally speaking while we may not always be able to predict exactly what rank a song would be, we can more often predict its quartile and we can more often than not predict within 1 quartile of its actual value.

```python
quart1count = 0
diffquart1 = list()
quart2count = 0
diffquart2 = list()
quart3count = 0
diffquart3 = list()
quart4count = 0
diffquart4 = list()
for q in range(len(y_test)):
    if y_test[q] == 1:
        quart1count += 1
        diffquart1.append(y_test[q]-pred[q])
    if y_test[q] == 2:
        quart2count += 1
        diffquart2.append(y_test[q]-pred[q])
    if y_test[q] == 3:
        quart3count += 1
        diffquart3.append(y_test[q]-pred[q])
    if y_test[q] == 4:
        quart4count += 1
        diffquart4.append(y_test[q]-pred[q])
accq1= 0
for h in diffquart1:
    if h == 0:
        accq1 += 1
print "Accurately Predicted as Quartile 1",(accq1/float(len(diffquart1)))*100, "%"
accq2= 0
for h in diffquart2:
    if h == 0:
        accq2 += 1
print "Accurately Predicted as Quartile 2",(accq2/float(len(diffquart2)))*100, "%"
accq3= 0
for h in diffquart3:
    if h == 0:
        accq3 += 1
print "Accurately Predicted as Quartile 3",(accq3/float(len(diffquart3)))*100, "%"
accq4= 0
for h in diffquart4:
    if h == 0:
        accq4 += 1
print "Accurately Predicted as Quartile 4",(accq4/float(len(diffquart4)))*100, "%"
```

So it seems that certain quartiles are predicted correctly more often than others. Songs in Quartile 1 are predicted correctly more often than songs in any other quartile, and songs in Quartile 3 are predicted correctly the least often.
